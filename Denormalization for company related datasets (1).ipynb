# Denormalization of company related datasets

To denormalize all the four company related datasets to one single collection, we first had to load all the 4 cleaned datasets and then merge them together

```python
import pandas as pd

# Load the first dataset
file_path_companies = 'cleaned_companies (1).csv'
companies_df = pd.read_csv(file_path_companies)

# Display the first few rows of the dataset
companies_df.head()
```

```python
# Load the second dataset
file_path_industries = 'modified_company_industries.csv'
industries_df = pd.read_csv(file_path_industries)

# Display the first few rows of the dataset
industries_df.head()
```

```python
# Load the third dataset
file_path_specialities = 'modified_company_specialities.csv'
specialities_df = pd.read_csv(file_path_specialities)

# Display the first few rows of the dataset
specialities_df.head()
```

```python
# Load the fourth dataset
file_path_employee_counts = 'modified_employee_counts.csv'
employee_counts_df = pd.read_csv(file_path_employee_counts)

# Display the first few rows of the dataset
employee_counts_df.head()
```

```python
# Merging the datasets on 'company_id'
merged_df = companies_df.merge(industries_df, on='company_id', how='left')
merged_df = merged_df.merge(specialities_df, on='company_id', how='left')
merged_df = merged_df.merge(employee_counts_df, on='company_id', how='left')

# Display the first few rows of the merged dataset
merged_df.head()
```
