```python
import pandas as pd
import pymongo
from bson import ObjectId
```

```python
# Load the CSV file
df = pd.read_csv('updated_denormalized_job_data.csv')

# Convert DataFrame to dictionary
data = df.to_dict(orient='records')

# Connect to MongoDB (update host and port as needed)
client = pymongo.MongoClient('mongodb://localhost:27017/')
db = client['lab2']
jobs_collection = db['job_postings']

# Transform and insert data
for record in data:
    # Convert 'company_id' to string (removing '.0')
    if 'job_id' in record and pd.notna(record['job_id']):
        record['job_id'] = str(record['job_id']).rstrip('.0')

    # Insert the record
    jobs_collection.insert_one(record)
```

```python
# Connect to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')
db = client['lab2']

# Fetch and print a few documents from the 'jobs' collection
for job in db.job_postings.find().limit(5):
    print(job)

# Count and print the number of documents in the 'jobs' collection
count = db.job_postings.count_documents({})
print("Total documents in the 'job_postings' collection:", count)
```
